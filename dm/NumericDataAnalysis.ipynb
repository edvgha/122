```python
import matplotlib.pyplot as plt
A1 = df_magic[0]
A2 = df_magic[1]
plt.scatter(A1, A2, color='red', marker='o', label='Point')
plt.show()
```

```python
import numpy as np
from scipy import stats

mu = A1.mean()
sigma = A1.std()
print(f'mean: {mu}')
print(f'std: {sigma}')

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
y = pd.DataFrame(0, index=np.arange(len(A1)), columns=[0])
plt.scatter(A1, y)
plt.show()
```
