```python
import pandas as pd

df_adult = pd.read_csv('adult.data',
                      header=None)

df_adult.columns = ['age', 'workclass', 'fnlwgt', 'education',
                   'education-num', 'marital-status', 'occupation',
                   'relationship', 'race', 'sex',
                   'capital-gain', 'capital-loss',
                   'hours-per-week', 'native-country', '<=50K']

df_adult.head()
```

```python
crosstab = pd.crosstab(df_adult['race'], df_adult['education'])
crosstab
```

```python
n = crosstab.sum().sum()
n_edu = crosstab.sum(axis=0)
n_race = crosstab.sum(axis=1)
```

```python
n_edu
```

```python
n_race
```

```python
import numpy as np
df = pd.DataFrame(np.outer(n_race, n_edu) / n, index = n_race.index, columns = n_edu.index)
df
```

```python
chi_square = (((crosstab - df) * (crosstab - df)) / df).sum().sum()
chi_square
```

```python
q = (df.shape[0] - 1) * (df.shape[1] - 1)
q
```

```python
from scipy import stats
CDF_chi_square = stats.chi2.cdf(chi_square, q)
CDF_chi_square
```

```python
p_value = 1 - CDF_chi_square
p_value
```
