# TANH

```python
import numpy as np


#6 layer each hiden size is 4096
dims = [4096] * 7 
hs = []

x = np.random.randn(1, dims[0])

for Din, Dout in zip(dims[:-1], dims[1:]):
    W = 0.1 * np.random.randn(Din, Dout)
    x = np.tanh(x.dot(W))
    hs.append(x)
    
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

fig, ax = plt.subplots(nrows=1, ncols=6)
ax[0].hist(np.squeeze(hs[0]))
ax[1].hist(np.squeeze(hs[1]))
ax[2].hist(np.squeeze(hs[2]))
ax[3].hist(np.squeeze(hs[3]))
ax[4].hist(np.squeeze(hs[4]))
ax[5].hist(np.squeeze(hs[5]))

plt.show()
```

```python
hs = []

x = np.random.randn(1, dims[0])

for Din, Dout in zip(dims[:-1], dims[1:]):
    W = np.random.randn(Din, Dout) / np.sqrt(Din)
    x = np.tanh(x.dot(W))
    hs.append(x)
    
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

fig, ax = plt.subplots(nrows=1, ncols=6)
ax[0].hist(np.squeeze(hs[0]))
ax[1].hist(np.squeeze(hs[1]))
ax[2].hist(np.squeeze(hs[2]))
ax[3].hist(np.squeeze(hs[3]))
ax[4].hist(np.squeeze(hs[4]))
ax[5].hist(np.squeeze(hs[5]))

plt.show()
```

# ReLU

```python
import numpy as np


#6 layer each hiden size is 4096
dims = [4096] * 7 
hs = []

x = np.random.randn(1, dims[0])

for Din, Dout in zip(dims[:-1], dims[1:]):
    W = np.random.randn(Din, Dout) / np.sqrt(Din)
    x = np.maximum(0, x.dot(W))
    hs.append(x)
    
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

fig, ax = plt.subplots(nrows=1, ncols=6)
ax[0].hist(np.squeeze(hs[0]))
ax[1].hist(np.squeeze(hs[1]))
ax[2].hist(np.squeeze(hs[2]))
ax[3].hist(np.squeeze(hs[3]))
ax[4].hist(np.squeeze(hs[4]))
ax[5].hist(np.squeeze(hs[5]))

plt.show()
```

```python
hs = []

x = np.random.randn(1, dims[0])

for Din, Dout in zip(dims[:-1], dims[1:]):
    W = np.random.randn(Din, Dout) / np.sqrt(2 / Din)
    x = np.maximum(0, x.dot(W))
    hs.append(x)
    
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

fig, ax = plt.subplots(nrows=1, ncols=6)
ax[0].hist(np.squeeze(hs[0]))
ax[1].hist(np.squeeze(hs[1]))
ax[2].hist(np.squeeze(hs[2]))
ax[3].hist(np.squeeze(hs[3]))
ax[4].hist(np.squeeze(hs[4]))
ax[5].hist(np.squeeze(hs[5]))

plt.show()
```
